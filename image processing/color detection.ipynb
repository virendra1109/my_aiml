```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
arr = np.ones((10,30,3),dtype="uint8")
arr
```

```python
arr [ : ,:10, :] = [255 , 0 , 0]
arr [ :, 10:20 ,: ] = [0,255,0]
arr [:,20:,:] = [0,0,255]
```

```python
plt.imshow(arr)
```

```python
plt.imsave("arr_img.png",arr)
```

```python
arr_g = cv2.cvtColor(arr,cv2.COLOR_RGB2GRAY)
plt.imshow(arr_g,cmap="gray")
```

```python
r = arr[:,:,2]
print(r.shape)
print(arr_g.shape)
plt.imshow(r,cmap="gray")
```
